# **Searching For Machine Learning and Data Mining Keyword**

```python
category_keyword_dict = {}
for i in range(keywords.shape[0]):
  if keywords.iloc[i][0] not in category_keyword_dict.keys():
    category_keyword_dict[keywords.iloc[i][0]] = [keywords.iloc[i][1]]
  else:
    category_keyword_dict[keywords.iloc[i][0]].append(keywords.iloc[i][1])
# category_keyword_dict['Machine learning & data mining']
ml_data_mining_keywords = []
for i in category_keyword_dict['Machine learning & data mining']:
  words = i.split()
  for j in words:
    if j not in ml_data_mining_keywords:
      ml_data_mining_keywords.append(j)
ml_data_mining_keywords
```

```python
author_keywords = pd.read_csv('/content/drive/MyDrive/Research/BCS2 Lab/Task 1 v2/author_keywords.csv')
author_keywords
```

```python
fullset['Index.Keywords']
```

```python
# ml_set => those index and index keywords present in custom keyword set 
ml_set = []
ml_index = []
for i in range(len(fullset['Index.Keywords'])):
  if fullset['Index.Keywords'][i] is not np.nan:
    words = fullset['Index.Keywords'][i].lower().split('; ')
    for j in words:
      if j not in ml_data_mining_keywords:
        ml_set.append(words)
        ml_index.append(i)
        break
len(ml_set)
```

```python
# ml_sources => source titles present in those indices
ml_sources = []
for i in ml_index:
  if fullset['Source.title'][i] is not np.nan and fullset['Source.title'][i] not in ml_sources:
    ml_sources.append(fullset['Source.title'][i])
```

```python
len(ml_sources)
```

```python
# ml_conf => conference titles present in those indices 
ml_conf = []
for i in ml_index:
  if fullset['Conference.name'][i] is not np.nan and fullset['Conference.name'][i] not in ml_conf:
    ml_conf.append(fullset['Conference.name'][i])
```

```python
len(ml_conf)
```

```python
# for more to the point keywords for searching in conference/source title 
ml_conf_keywords = ['mining', 'machine', 'learning']
```

```python
# more accurate list of conferences
ml_conference = []
for i in ml_conf:
  words = i.lower().split()
  for j in words:
    if j in ml_conf_keywords and i not in ml_conference:
      ml_conference.append(i)
len(ml_conference)
```

```python
# more accurate list of sources
ml_source = []
for i in ml_sources:
  words = i.lower().split()
  for j in words:
    if j in ml_conf_keywords and i not in ml_source:
      ml_source.append(i)
len(ml_source)
```

# **Further Playing with Keywords**

```python
all_keywords = []
indices = []
keyword_conf_source_dict ={}
for i in range(fullset.shape[0]):
  if fullset['Conference.name'][i] in ml_conference:
    indices.append(i)
    if fullset['Author.Keywords'][i] is not np.nan:
      words = fullset['Author.Keywords'][i].lower().split('; ')
      for j in words:
        if j not in all_keywords: 
          all_keywords.append(j)
          if fullset['Conference.name'][i] not in keyword_conf_source_dict.keys():
            keyword_conf_source_dict[fullset['Conference.name'][i]] = [j]
          else:
            keyword_conf_source_dict[fullset['Conference.name'][i]].append(j)
    if fullset['Index.Keywords'][i] is not np.nan:
      words2 = fullset['Index.Keywords'][i].lower().split('; ')
      for j in words2:
        if j not in all_keywords:
          all_keywords.append(j)
          if fullset['Conference.name'][i] not in keyword_conf_source_dict.keys():
            keyword_conf_source_dict[fullset['Conference.name'][i]] = [j]
          else:
            keyword_conf_source_dict[fullset['Conference.name'][i]].append(j)
print(len(all_keywords), len(indices), len(list(keyword_conf_source_dict.keys())))
```

```python
for i in range(fullset.shape[0]):
  if fullset['Source.title'][i] in ml_source:
    if i not in indices:
      indices.append(i)
      if fullset['Author.Keywords'][i] is not np.nan:
        words = fullset['Author.Keywords'][i].lower().split('; ')
        for j in words:
          if j not in all_keywords: 
            all_keywords.append(j)
            if fullset['Source.title'][i] not in keyword_conf_source_dict.keys():
              keyword_conf_source_dict[fullset['Source.title'][i]] = [j]
            else:
              keyword_conf_source_dict[fullset['Source.title'][i]].append(j)
      if fullset['Index.Keywords'][i] is not np.nan:
        words2 = fullset['Index.Keywords'][i].lower().split('; ')
        for j in words2:
          if j not in all_keywords:
            all_keywords.append(j)
            if fullset['Source.title'][i] not in keyword_conf_source_dict.keys():
              keyword_conf_source_dict[fullset['Source.title'][i]] = [j]
            else:
              keyword_conf_source_dict[fullset['Source.title'][i]].append(j)
print(len(all_keywords), len(indices), len(list(keyword_conf_source_dict.keys())))
```

```python
source = []
keyword = []
for i in keyword_conf_source_dict.keys():
  for j in keyword_conf_source_dict[i]:
    source.append(i)
    keyword.append(j)
print(len(source))
print(len(keyword))
```
